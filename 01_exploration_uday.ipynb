```python
import pandas as pd
import plotly as plt
import plotly.express as px
import glob
import os
```

```python
df = pd.read_csv('raw_data.csv');
print(df)
```

```python
cols = ['ClosePrice', 'LivingArea', 'BedroomsTotal', 'BathroomsTotalInteger', 'LotSizeAcres'];
eda_df = df[cols];
eda_df['Group'] = ((df['PropertyType'] == 'Residential') & (df['PropertySubType'] == 'SingleFamilyResidence')).map({True: 'Single Family Residential', False: 'Other'});
print(eda_df);
```

```python
fig = px.histogram(eda_df, x="ClosePrice", color="Group", range_x=[0, 5000000], range_y=[0, 30000], title="Close Price Distribution", labels={"ClosePrice": "Close Price"});
fig.show();
```

```python
#Number of bins is scaled too low, please change to density plot
fig = px.histogram(eda_df, x="LivingArea", range_x=[0, 10000], range_y=[0, 600000], color="Group");
fig.show();
```

```python
fig = px.histogram(eda_df, x="BedroomsTotal", color="Group", range_x = [0, 10], title="# of Bedrooms Distribution", labels={"BedroomsTotal": "# of Bedrooms"});
fig.show();
```

```python
fig = px.histogram(eda_df, x="BathroomsTotalInteger", color="Group", range_x = [0, 10], title="# of Bathrooms Distribution", labels={"BathroomsTotalInteger":"# of Bathrooms"});
fig.show();
```

```python
#Change to density plot as well
fig = px.histogram(eda_df, x="LotSizeAcres", color="Group", range_x=[0, 20], title="Lot Size Distribution", labels={'LotSizeAcres':"Lot Size (Acres)"});
fig.show();
```

```python
fig = px.scatter(
    eda_df, x='ClosePrice', y='LivingArea', color='Group', 
    trendline="ols",  # "ols" = Ordinary Least Squares
    title="Close Price vs Living Area", range_x=[0, 50000000], range_y=[0, 100000],
    labels={"ClosePrice":"Close Price", "LivingArea":"Living Area"}
)

fig.show()
```
